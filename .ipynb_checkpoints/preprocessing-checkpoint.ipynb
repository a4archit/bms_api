```python

# Set isbn13 as index (optional)
df.set_index('isbn13', inplace=True)
```

```python
df.fillna('', inplace=True)
```

```python
total_cols = len(df.columns)
cols = df.columns
```

```python


# Convert each row to a list and then to dictionary
json_data = df.apply(lambda row: {cols[i]: row.tolist()[i] for i in range(total_cols)}, axis=1).to_dict()

# print(type(json_data))

# Save to JSON file
# df.to_json("books_data.json", orient="records", indent=4)

import json
with open("books_data.json", "w", encoding='utf-8') as f:
    print(json_data)
    json.dump(json_data, f, indent=4)
```
